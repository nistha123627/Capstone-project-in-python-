```python
import pandas as pd

df1 = pd.read_csv("my_data.csv")
df2 = pd.read_csv("filename.csv")
df3 = pd.read_csv("employee_data.csv")

# Display the dataframes
print("DF1:")
display(df1)

print("DF2:")
display(df2)

print("DF3:")
display(df3)
```

```python
Final = df1.merge(df2, on="id", how="inner").merge(df3, on="id", how="inner")
```

```python
Final = df1.merge(df2, on="id", how="inner").merge(df3, on="id", how="inner")

display(Final)
```

```python
Final["bonus"] = Final.apply(
    lambda row: row["cost"] * 0.05 if row["status"].strip().lower() == "finished" else 0,
    axis=1
)
display(Final)
```

```python
df = pd.DataFrame(Final)

# Step 1: Demote designation level by 1 for employees with failed projects
df.loc[df['status'].str.lower() == 'failed', 'Designation level'] -= 1

# Step 2: Delete employees whose designation level is above 4
df = df[df['Designation level'] <= 4]

# Display the updated DataFrame
print(df)
```

```python
df = pd.DataFrame(Final)

# Step 1: Add "Mr." or "Mrs." prefix based on Gender
df['First Name'] = df.apply(lambda x: "Mr. " + x['First Name'] if x['Gender'] == 'M' else "Mrs. " + x['First Name'], axis=1)

# Step 2: Drop the Gender column
df = df.drop(columns=['Gender'])

# Display the updated DataFrame
print(df)
```

```python
df = pd.DataFrame(Final)

# Promote designation level by 1 for employees whose age > 29
df['Designation level'] = df.apply(lambda x: x['Designation level'] + 1 if x['Age'] > 29 else x['Designation level'], axis=1)

# Display the updated DataFrame
print(df)
```

```python
df = pd.DataFrame(Final)

# Group by ID and First Name and sum the cost
TotalProjCost = df.groupby(['id','First Name'], as_index=False)['cost'].sum()

# Rename columns
TotalProjCost = TotalProjCost.rename(columns={'cost': 'Total cost'})

# Display the new DataFrame
print(TotalProjCost)
```

```python
df = pd.DataFrame(Final)

# Filter rows where City contains the letter "o" (case-insensitive)
filtered_df = df[df['City'].str.contains('o', case=False)]

# Display the filtered employee details
print(filtered_df)
```
